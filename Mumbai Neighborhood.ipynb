## find out longitude latitue of this areas using geopy

```python
from geopy.geocoders import Nominatim
from pandas.io.json import json_normalize
```

```python
def loca(address):
    geolocator = Nominatim(user_agent = 'foursquare_agent')
    location = geolocator.geocode(address)
    try:
        return(location.latitude, location.longitude)
    except:
        return None, None 
```

```python
list2 = []
for neighborhood, ID in zip(neigh['Neighborhood'], neigh['ID']):
    address= neighborhood + ' Toronto'
    lat, lng = loca(address)
    list2.append([ID, lat, lng])
```

```python
latlong = pd.DataFrame(list2)
latlong.columns = ['ID','Latitude','Longitude']
latlong.head()
```

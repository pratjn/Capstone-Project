### loop all neighborhoods to find nearby venues

```python
def getvenues (names, latitude, longitude, rad = 500):
    venuelist = []
    for name, lat, lng in zip(names, latitude, longitude):
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius ={}&limit ={}'.format(CLIENT_ID,CLIENT_SECRET,VERSION,lat,lng,rad,LIMIT)
        results = requests.get(url).json()
        data = json_normalize(results['response']['groups'][0]['items'])
        for i in range(0,len(data)):
            venuelist.append([name,lat,lng,data['venue.name'][i],data['venue.categories'][i][0]['name'],data['venue.location.lat'][i],data['venue.location.lng'][i]])
    print (venuelist)
    return (venuelist)
```

## create a data frame that has top three venues in each neighborhood and there frequency 

```python
## lets find out number of venues in each neighborhood
venues.groupby(['Neighborhood']).count()
```

```python
#neigh = list(toronto.Neighborhood)[38]
temp = venues[venues['Neighborhood']==neigh]['venue type']
freqtop = list(temp.value_counts())[0:3]
top = list(temp.value_counts().index)[0:3]
dfven=[]
for neigh in list(toronto.Neighborhood):
    temp = venues[venues['Neighborhood']==neigh]['venue type']
    freqtop = list(temp.value_counts())[0:3]
    top = list(temp.value_counts().index)[0:3]
    dfven.append([neigh]+top+freqtop)
dfven
    
```

```python
df2  = pd.DataFrame(dfven)
df2.columns = ['Neighborhood','1st most common venue', '2nd most common venue','3rd most common venue','frequency of first venue ','frequesncy of 2nd venue','frequency of third venue']
```

```python
df2.head()
```

## use kmeans for clustering above data frame 

```python
from sklearn.cluster import KMeans
km = KMeans(init = "k-means++", n_clusters = 5, n_init = 12)
```

```python
dfcluster= df2.iloc[:,4:7]
dfcluster.head()
```

```python
km.fit(dfcluster)
```

```python
label = pd.DataFrame(km.labels_)
label.columns = ['Label']
```

```python
df3= df2.join(label)
```

```python
df3.head()
```

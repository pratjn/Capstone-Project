# Introduction

This study was done to identify best area to open a restaurent in toronto

Best neighborhood will be the place which has high population density, High median income of people, low shelter cost and less competetors

# Data

data for this study was scraped from open source census data on toronto website 
link : https://www.toronto.ca/city-government/data-research-maps/neighbourhoods-communities/neighbourhood-profiles/
below are data points collected 
01. Neighborhoods in Toronto and their latitude longitude 
02. Average population in each neighborhood 
03. Median income in each neighborhood 
04. number of competetors in each neighborhood

# Methodology 

## 01. Data accumulation 

List of neighborhood in toront was scraped from open data source of toronto. There longitude and latitude where then updated in data frame with the help of foursquare api.

Population and income data for neighborhoods were imported from https://open.toronto.ca/dataset/wellbeing-toronto-demographics/ available in excel format

number of similar outlets also collected with the help of foursquare api and visulized on the map 

## 02. Data wrangling

few latitude and longitude was not found by foursquare api. there were manually put in list and appended in dataframe using nearby location of areas using google maps

Data cleaning was done and orientation of data with correct data type

data was visualised with the help of folium maps



## 03. Clustering

01. **K means Clustering** was used to cluster this 140 neighborhood on the basis of Population and Median Income to identify high potential consumption areas. K means clustering was used because we wanted to classify all out neighborhoods in fixed number of clusters on the basis of their high potential to consumption

2. after clustering, most favourable class was identified on the basis of Index = (avg. Population* avg median Income)/ ( avg. cost on real estate in that area)


3. now all competetors restaurents in each neighborhood with highest index is taken out with the help of explore option in foursquare api 


We can see from above that Nighborhoods **Rouge** and **Waterfront Communities-The Island** has least number of competetors and have high potential volume of sales as well as low cost of opening a restaurent this two locations will be best suited for opening a restaurent
